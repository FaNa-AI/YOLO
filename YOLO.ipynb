```python
from ultralytics import YOLO

# Load a pretrained YOLOv8n model
# 'n' stands for nano, which is the smallest and fastest version
model = YOLO('yolov8n.pt')

# Train the model on the coco128 dataset for a few epochs
# epochs: Number of training epochs (iterations over the entire dataset)
# imgsz: Input image size
# data: Path to the dataset configuration file (Ultralytics provides coco128.yaml)
print("Starting model training...")
results = model.train(data='coco128.yaml', epochs=5, imgsz=640)

print("Training finished!")
```

```python
from ultralytics import YOLO

# Load a pretrained YOLOv8n model
# 'n' stands for nano, which is the smallest and fastest version
model = YOLO('yolov8n.pt')

# Train the model on the coco128 dataset for a few epochs
# epochs: Number of training epochs (iterations over the entire dataset)
# imgsz: Input image size
# data: Path to the dataset configuration file (Ultralytics provides coco128.yaml)
print("Starting model training...")
results = model.train(data='coco128.yaml', epochs=5, imgsz=640)

print("Training finished!")
```

```python
# Use the trained model to make predictions on an image
# For this example, let's use a sample image that Ultralytics provides
# or you can upload your own image to Colab.
# If you upload your own, make sure to change 'path/to/your/image.jpg'

# You can get a sample image from the internet or use one from Colab's default paths
# For example, let's use a common sample image path if available
# Or download one for demonstration:
!wget -q https://ultralytics.com/images/bus.jpg -O bus.jpg

print("Running prediction on bus.jpg...")
# Perform inference on the image
results = model.predict(source='bus.jpg', conf=0.25) # conf is confidence threshold

# Show the results (this will display the image with bounding boxes)
# Iterating through results object will show individual prediction objects
for r in results:
    # r.show() will display the image in Colab output
    r.show()
    # r.save(filename='prediction.jpg') # You can also save the image
```
